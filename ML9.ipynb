```
import matplotlib.pyplot as plt
import numpy as np

file_data   = "/gdrive/My Drive/ML_assignment/mnist.csv"
handle_file = open(file_data, "r")
data        = handle_file.readlines()
handle_file.close()

size_row    = 28    # height of the image
size_col    = 28    # width of the image

num_image   = len(data)
count       = 0     # count for the number of images

#
# normalize the values of the input data to be [0, 1]
#
def normalize(data):

    data_normalized = (data - min(data)) / (max(data) - min(data))

    return(data_normalized)

#
# example of distance function between two vectors x and y
#
def distance(x, y):

    d = (x - y) ** 2
    s = np.sum(d)
    # r = np.sqrt(s)

    return(s)

#
# make a matrix each column of which represents an images in a vector form
#
list_image  = np.empty((size_row * size_col, num_image), dtype=float)
list_label  = np.empty(num_image, dtype=int)

for line in data:

    line_data   = line.split(',')
    label       = line_data[0]
    im_vector   = np.asfarray(line_data[1:])
    im_vector   = normalize(im_vector)

    list_label[count]       = label
    list_image[:, count]    = im_vector

    count += 1

#
# plot first 150 images out of 10,000 with their labels
#
f1 = plt.figure(1)

for i in range(150):

    label       = list_label[i]
    im_vector   = list_image[:, i]
    im_matrix   = im_vector.reshape((size_row, size_col))

    plt.subplot(10, 15, i+1)
    plt.title(label)
    plt.imshow(im_matrix, cmap='Greys', interpolation='None')

    frame   = plt.gca()
    frame.axes.get_xaxis().set_visible(False)
    frame.axes.get_yaxis().set_visible(False)


#plt.show()

#
# plot the average image of all the images for each digit
#
f2 = plt.figure(2)

im_average  = np.zeros((size_row * size_col, 10), dtype=float)
im_count    = np.zeros(10, dtype=int)

for i in range(num_image):

    im_average[:, list_label[i]] += list_image[:, i]
    im_count[list_label[i]] += 1

for i in range(10):

    im_average[:, i] /= im_count[i]

    plt.subplot(2, 5, i+1)
    plt.title(i)
    plt.imshow(im_average[:,i].reshape((size_row, size_col)), cmap='Greys', interpolation='None')

    frame   = plt.gca()
    frame.axes.get_xaxis().set_visible(False)
    frame.axes.get_yaxis().set_visible(False)

plt.show()
```

```
import math
# 1. 세타 값 initialization
mu, sigma = 0, 0.1
u=np.random.normal(mu,sigma,(196,785))
v=np.random.normal(mu,sigma,(49,197))
w=np.random.normal(mu,sigma,(10,50))

ut=np.random.normal(mu,sigma,(196,785))
vt=np.random.normal(mu,sigma,(49,197))
wt=np.random.normal(mu,sigma,(10,50))

#2. input 값 정리
train_x=list_image[:,0:6000]
train_l=list_label[0:6000]

test_x=list_image[:,6000:]
test_l=list_label[6000:]

#bias 값 추가
temp1=np.ones((6000), dtype=float)
temp2=np.ones((4000), dtype=float)

train_x=np.insert(train_x,0,temp1,axis=0)
test_x=np.insert(test_x,0,temp2,axis=0)

#plot위한 list
J_train=[]
accurate_train=[]
steps=[]

J_test=[]
accurate_test=[]

#sigmoid 함수 이후 값
y_s = np.zeros((197, 6000), dtype=float)
z_s = np.zeros((50, 6000), dtype=float)
h_s = np.zeros((10, 6000), dtype=float)

y_st = np.zeros((197, 4000), dtype=float)
z_st = np.zeros((50, 4000), dtype=float)
h_st = np.zeros((10, 4000), dtype=float)

#label위한 diagonal 행렬
l=np.identity(10)

#bias 값 추가(sigmoid layer에만)
for i in range(0,6000):
  y_s[0][i]=1
  z_s[0][i]=1
  if(i<4000):
    y_st[0][i]=1
    z_st[0][i]=1

for c in range(0,1000):
  steps.append(c)

  # 3. J 구하기
  #3-1 y랑 sigmoid y
  y=np.dot(u,train_x)
  yt=np.dot(ut,test_x)
  for a in range(1,197):
    for b in range(0,6000):
      y_s[a][b]=1/(1+math.exp(-y[a-1][b]))
      if(b<4000):
        y_st[a][b]=1/(1+math.exp(-yt[a-1][b]))

  #3-2 z 랑 sigmoid z
  z=np.dot(v,y_s)
  zt=np.dot(vt,y_st)
  for a in range(1,50):
    for b in range(0, 6000):
      z_s[a][b]=1/(1+math.exp(-z[a-1][b]))
      if(b<4000):
        z_st[a][b]=1/(1+math.exp(-zt[a-1][b]))

  #3-3 H랑 sigmoid h
  h=np.dot(w,z_s)
  ht=np.dot(wt,z_st)
  for a in range(0,10):
    for b in range(0,6000):
      h_s[a][b]=1/(1+math.exp(-h[a][b]))
      if(b<4000):
         h_st[a][b]=1/(1+math.exp(-ht[a][b]))

  #5. gradient descent 하기, J 구하기, accurate 세기
  J_sum=0
  p=0.5
  h_s_t=np.transpose(h_s)
  z_s_t=np.transpose(z_s[1:,:])
  y_s_t=np.transpose(y_s[1:,:])
  x_s_t=np.transpose(train_x[1:,:])

  w_t=w[:,1:]
  v_t=v[:,1:]
  u_t=u[:,1:]
  
  one_z=np.transpose(np.ones((49), dtype=float))
  one_y=np.transpose(np.ones((196), dtype=float))

  delta4=np.zeros((49,10), dtype=float)
  delta3=np.zeros((196,49), dtype=float)
  delta2=np.zeros((784,196), dtype=float)

  ################test
  Jt_sum=0

  h_s_t2=np.transpose(h_st)
  z_s_t2=np.transpose(z_st[1:,:])
  y_s_t2=np.transpose(y_st[1:,:])
  x_s_t2=np.transpose(test_x[1:,:])

  w_t2=wt[:,1:]
  v_t2=vt[:,1:]
  u_t2=ut[:,1:]
  
  one_z2=np.transpose(np.ones((49), dtype=float))
  one_y2=np.transpose(np.ones((196), dtype=float))

  delta42=np.zeros((49,10), dtype=float)
  delta32=np.zeros((196,49), dtype=float)
  delta22=np.zeros((784,196), dtype=float)
  ########################
  result_test=np.empty((4000), dtype=int)
  true_table_test=np.empty((4000), dtype=int)
  false_table_test=np.empty((4000), dtype=int)

  false_count_test=0
  count_test=0
  count=0
  for i in range(0,6000):
    result=np.argmax(h_s[:,i])
    if(result==train_l[i]):
      count=count+1

    lable=train_l[i]
    temp4= h_s_t[i]-l[lable]
    temp3=np.multiply( np.dot( np.transpose( w_t ) , temp4 ) , np.multiply(z_s_t[i] , one_z - z_s_t[i]))
    temp2=np.multiply( np.dot( np.transpose( v_t ) , temp3 ) , np.multiply(y_s_t[i] , one_y - y_s_t[i]))
    delta4=delta4 + np.dot( np.transpose(z_s_t[i][np.newaxis]), temp4[np.newaxis])
    delta3=delta3+ np.dot( np.transpose(y_s_t[i][np.newaxis]), temp3[np.newaxis] )
    delta2=delta2+ np.dot(np.transpose(x_s_t[i][np.newaxis]), temp2[np.newaxis])

    for k in range(0,10):
        J_sum=J_sum+(-l[lable][k]*math.log(h_s[k][i])-(1-l[lable][k])*math.log(1-h_s[k][i]))
    
    #######################test
    
    if i<4000:
      result_test[i]=np.argmax(h_st[:,i])
      if(result_test[i]==test_l[i]):
        true_table_test[count_test]=i
        count_test=count_test+1
      else:
        false_table_test[false_count_test]=i
        false_count_test=false_count_test+1

      lable_test=test_l[i]
      temp4t= h_s_t2[i]-l[lable_test]
      temp3t=np.multiply( np.dot( np.transpose( w_t2 ) , temp4t ) , np.multiply(z_s_t2[i] , one_z2 - z_s_t2[i]))
      temp2t=np.multiply( np.dot( np.transpose( v_t2 ) , temp3t ) , np.multiply(y_s_t2[i] , one_y2 - y_s_t2[i]))
      delta42=delta42 + np.dot( np.transpose(z_s_t2[i][np.newaxis]), temp4t[np.newaxis])
      delta32=delta32 + np.dot( np.transpose(y_s_t2[i][np.newaxis]), temp3t[np.newaxis] )
      delta22=delta22 + np.dot(np.transpose(x_s_t2[i][np.newaxis]), temp2t[np.newaxis])

      for k in range(0,10):
          Jt_sum=Jt_sum+(-l[lable_test][k]*math.log(h_st[k][i])-(1-l[lable_test][k])*math.log(1-h_st[k][i]))
    ###########################

  accuracy=count/6000
  accurate_train.append(accuracy)

  delta4=delta4/6000
  delta3=delta3/6000
  delta2=delta2/6000

  J_sum=J_sum/6000
  J_train.append(J_sum)

  ####################test
  accuracy_test=count_test/4000
  accurate_test.append(accuracy_test)
  delta42=delta42/4000
  delta32=delta32/4000
  delta22=delta22/4000
  Jt_sum=Jt_sum/4000
  J_test.append(Jt_sum)
  ######################

  for a in range(0,196):
    for b in range(1,785):
      if a<10 and b<50:
        w[a][b]=w[a][b]-p*(np.transpose(delta4)[a][b-1])
        wt[a][b]=wt[a][b]-p*(np.transpose(delta42)[a][b-1])
      if a<49 and b<197:
        v[a][b]=v[a][b]-p*np.transpose(delta3)[a][b-1]
        vt[a][b]=vt[a][b]-p*np.transpose(delta32)[a][b-1]
      u[a][b]=u[a][b]-p*np.transpose(delta2)[a][b-1]
      ut[a][b]=ut[a][b]-p*np.transpose(delta22)[a][b-1]

  print(c,"train:::acccuracy  ",accuracy,"   J   ",J_sum,"::::: test accuracy",accuracy_test,"   J   ",Jt_sum)


f1 = plt.figure(1)
plt.scatter(steps,J_train,c="red",s=0.3)
plt.scatter(steps,J_test,c="blue",s=0.3)

f2 = plt.figure(2)
plt.scatter(steps,accurate_train,c="red",s=0.3)
plt.scatter(steps,accurate_test,c="blue",s=0.3)
```
