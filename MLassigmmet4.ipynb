**1. Plot the estimated parameters using the training dataset **

- plot the estimated parameters {(θ0,θ1,θ2,θ3)} at every iteration of gradient descent until convergence
- the colors for the parameters {(θ0,θ1,θ2,θ3)} should be black, red, green, blue, respectively
- the optimization is performed using the training dataset ('data_train.csv')

```
import numpy as np
import csv
import matplotlib.pyplot as plt
import random

#test& train data 불러오기
with open('/gdrive/My Drive/ML_assignment/data_test.csv', newline='') as myfile:
    reader  = csv.reader(myfile, delimiter=',')
    test_data=[]
    for i in reader:
        T=[float(i[0]),float(i[1]),float(i[2]),float(i[3])]
        test_data.append(T)
test_len=len(test_data)

with open('/gdrive/My Drive/ML_assignment/data_train.csv', newline='') as myfile:
    reader  = csv.reader(myfile, delimiter=',')
    train_data=[]
    Y=[]
    for i in reader:
        T=[float(i[0]),float(i[1]),float(i[2]),float(i[3])]
        train_data.append(T)
print(train_data)        
train_len=len(train_data)
```

```
#gradient descent
#J=1/2mΣ(theta(t)v(i)-h(i))

p=0.00001

d=5
dx=5
dy=5
dz=5

step=0

theta0=0
theta1=0
theta2=0
theta3=2
theta=np.array([theta0,theta1,theta2,theta3])

#theta 값들의 vector

while(((-0.002>d) or (d>0.002)) and ((-0.002>dx) or (dx>0.002)) and ((-0.002>dy) or (dy>0.002)) and ((-0.002>dz) or (dz>0.002))):
  step=step+1

  s=0
  sx=0
  sy=0
  sz=0

  for i in range(0,300):
    #데이터값들을 불러옴
    x=train_data[i][0]
    y=train_data[i][1]
    z=train_data[i][2]
    yy=train_data[i][3]
  
    #데이터값들에 대한 vector
    vector=np.array([1,x,y,z])
    
    #시그마 계산
    h=np.dot(np.transpose(theta),vector)-yy

    s=h*1+s
    sx=h*x+sx
    sy=h*y+sy
    sz=h*z+sz

  #델타 값 계산
  d=s/train_len
  dx=sx/train_len
  dy=sy/train_len
  dz=sz/train_len

  #세타 값 업데이트
  theta0=theta0-p*d
  theta1=theta1-p*dx
  theta2=theta2-p*dy
  theta3=theta3-p*dz

  plt.scatter(step,theta0,c="black",s=2)
  plt.scatter(step,theta1,c="red",s=2)
  plt.scatter(step,theta2,c="green",s=2)
  plt.scatter(step,theta3,c="blue",s=2)
  
  theta=np.array([theta0,theta1,theta2,theta3])
```

** 2. Plot the training error using the training dataset**

- plot the training error J(θ0,θ1,θ2,θ3)at every iteration of gradient descent until convergence (in blue color)

```
#gradient descent
#J=1/2mΣ(theta(t)v(i)-h(i))

p=0.0000001
d=100
dx=100
dy=100
dz=100
step=0
theta0=0
theta1=0
theta2=0
theta3=0
theta=np.array([theta0,theta1,theta2,theta3])

while((-0.1>d) or (d>0.1)) and ((-0.1>dx) or (dx>0.1)) and ((-0.1>dy) or (dy>0.1)) and ((-0.1>dz) or (dz>0.1)):
  step=step+1

  s=0
  sx=0
  sy=0
  sz=0
  s2=0

  for i in range(0,train_len):
    x=train_data[i][0]
    y=train_data[i][1]
    z=train_data[i][2]
    yy=train_data[i][3]

    vector=np.array([1,x,y,z])

    s=(sum(np.transpose(theta)*vector)-yy)+s
    sx=(sum(np.transpose(theta)*vector)-yy)*x+sx
    sy=(sum(np.transpose(theta)*vector)-yy)*y+sy
    sz=(sum(np.transpose(theta)*vector)-yy)*z+sz

    #J값 계산을 위해 
    s2=((sum(np.transpose(theta)*vector)-yy)**2)+s2

  d=s/train_len
  dx=sx/train_len
  dy=sy/train_len
  dz=sz/train_len

  theta0=theta0-p*d
  theta1=theta1-p*dx
  theta2=theta2-p*dy
  theta3=theta3-p*dz

  theta=np.array([theta0,theta1,theta2,theta3])
  J=s2/(2*train_len)
  plt.scatter(step,J,c="blue",s=1)
  
```

**3. Plot the testing error using the testing dataset at every iteration of gradient descent until convergence**

- plot the testing error J(θ0,θ1,θ2,θ3)at every iteration of gradient descent until convergence (in red color)

```
#gradient descent
#J=1/2mΣ(theta(t)v(i)-h(i))
p=0.0000001
d=100
dx=100
dy=100
dz=100
step=0
theta0=0
theta1=0
theta2=0
theta3=0

theta=np.array([theta0,theta1,theta2,theta3])

while((-0.1>d) or (d>0.1)) and ((-0.1>dx) or (dx>0.1)) and ((-0.1>dy) or (dy>0.1)) and ((-0.1>dz) or (dz>0.1)):
  step=step+1
  s=0
  sx=0
  sy=0
  sz=0
  s2=0
  test2=0 #test data 관련 변수

  for i in range(0,train_len):
    x=train_data[i][0]
    y=train_data[i][1]
    z=train_data[i][2]
    yy=train_data[i][3]

    #test data 받기
    xt=test_data[i][0]
    yt=test_data[i][1]
    zt=test_data[i][2]
    yyt=test_data[i][3]

    vector=np.array([1,x,y,z])
    test_vector=np.array([1,xt,yt,zt])
    
    s=(sum(np.transpose(theta)*vector)-yy)+s
    sx=(sum(np.transpose(theta)*vector)-yy)*x+sx
    sy=(sum(np.transpose(theta)*vector)-yy)*y+sy
    sz=(sum(np.transpose(theta)*vector)-yy)*z+sz

    #이번 step의 theta 벡터에 따른 cost를 계산하기 위해 (h(theta)-y)^2를 계산해서 다 더해놓음
    test2=((sum(np.transpose(theta)*test_vector)-yyt)**2)+test2

  d=s/train_len
  dx=sx/train_len
  dy=sy/train_len
  dz=sz/train_len

  theta0=theta0-p*d
  theta1=theta1-p*dx
  theta2=theta2-p*dy
  theta3=theta3-p*dz

  theta=np.array([theta0,theta1,theta2,theta3])
  
  #이번 step의 theta 벡터에 따른 test 값의 cost 계산
  Jtest=test2/(2*test_len)

  plt.scatter(step,Jtest,c="red",s=1)
  
```
